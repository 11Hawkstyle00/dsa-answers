```python
import time
import numpy as np
import matplotlib.pyplot as plt

N = 2
n_start = 1
n_end = 10**5 * N  
step = 100 * N   
num_runs = 5       

print("Начало эксперимента для Задания 2...")
print("Внимание: из-за O(n^3) сложности эксперимент может занять много времени!")
print("Для больших n время выполнения будет очень большим")


matrix_n_values = [1, 2, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
average_times_matmul = []

for i, n in enumerate(matrix_n_values):
    print(f"Точка {i+1}/{len(matrix_n_values)}, n={n}")
    
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    
    times_matmul = []
    
    for run in range(num_runs):
        start_time = time.perf_counter()
        
        C = np.zeros((n, n))
        for i in range(n):
            for j in range(n):
                sum_val = 0.0
                for k in range(n):
                    sum_val += A[i, k] * B[k, j]
                C[i, j] = sum_val
        
        end_time = time.perf_counter()
        times_matmul.append(end_time - start_time)
    
    average_times_matmul.append(sum(times_matmul) / num_runs)

print("Эксперимент для Задания 2 завершен!")

plt.figure(figsize=(12, 8))
plt.plot(matrix_n_values, average_times_matmul, 'o-', color='red', 
         label='Обычное матричное умножение (теор. O(n^3))', markersize=6, linewidth=2)

plt.xlabel('Размерность матрицы (n)')
plt.ylabel('Среднее время выполнения (сек)')
plt.title('Задание 2: Зависимость времени перемножения матриц от их размера\n(Для демонстрации O(n^3) сложности)')
plt.legend()
plt.grid(True, which="both", ls="--")
# plt.yscale('log')
# plt.xscale('log')
plt.tight_layout()
plt.show()
```
